```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read in the player data and combine
player1_df = pd.read_csv("PlayerStats_clean_2011-15.csv")
player2_df = pd.read_csv("PlayerStats_clean_2016.csv")

frames = [player1_df, player2_df]

player_df = pd.concat(frames)

games_list = player_df.groupby('Player').size().tolist()

# construct the list of players
players_list = player_df["Player"].values.tolist()
players_list = set(players_list)

plt.hist(games_list, bins=[0, 80, 160, 240, 320, 400, 480, 560, 600])
plt.title("Distribution of Games Played")
plt.xlabel("Number of Games")
plt.ylabel("Number of Players")
plt.show()
```

```python
games_list = player_df.groupby('Type').size().tolist()
player_types = [1, 2, 3, 4, 5]
width_list = [0.80 / 2] * 5


fig = plt.figure()
ax = plt.subplot(111)
ax.bar(player_types, games_list, width=0.8, align='center')
ax.set_xticks(player_types)
ax.set_xticklabels(['1', '2', '3', '4', '5'])
plt.title("Number of Games Per Player Type")
plt.xlabel("Player Type")
plt.ylabel("Number of Games")
plt.show()

print games_list
```
